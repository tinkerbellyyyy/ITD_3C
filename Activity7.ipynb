```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import folium
```

```python
# Load the dataset with the correct encoding
df = pd.read_csv('Datasets\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df.columns
```

<h1>
Filtering Data</h1>

```python
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']
filter3 = filter2[filter2['nkill'] >= 10]
filter4 = filter3[filter3['iyear'] >= 2000]
```

```python
filter4
```

<h1>Data Aggregation</h1>

```python
ME_terror = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')
ME_terror
```

<h1>Visualization of GeoSpatial using FOLIUM

```python
import folium
from folium.plugins import HeatMap
```

```python
heat_df = ME_terror[['longitude', 'latitude', 'incident_count']].values.tolist()
heat_df
```

```python
my_map = folium.Map(location=[30, 40], zoom_start=4)
```

```python
HeatMap(heat_df).add_to(my_map)
my_map
```

<h1>
Regional Analysis

```python
# Handle missing values for longitude, latitude, or region column
df = df.dropna(subset=['longitude', 'latitude', 'region_txt'])

# Group by longitude, latitude, and region, and calculate the count of incidents
regional_analysis = df.groupby(['longitude', 'latitude', 'region_txt']).size().reset_index(name='incident_count')

# Convert the data into a list format for use in the heatmap
heatmap_data_region = regional_analysis[['longitude', 'latitude', 'incident_count']].values.tolist()

# Create a base map centered on a general global location (adjust as needed for your dataset)
region_map = folium.Map(location=[20, 0], zoom_start=2)  # Global view

# Add the heatmap layer to the map
HeatMap(heatmap_data_region).add_to(region_map)

# Display the map
region_map
```

<h1>Crime/Attack Analysis

```python
# Load the dataset with the correct encoding
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')

# Apply filters to focus on Europe
europe_filter = df[df['region_txt'] == 'Western Europe']  # Filter for Western Europe as per the dataset classification

# Group by longitude and latitude and calculate the number of incidents at each location
europe_crime = europe_filter.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')

# Convert the data into a list format for use in the heatmap
heatmap_data_europe = europe_crime[['longitude', 'latitude', 'incident_count']].values.tolist()

# Create a base map centered on Europe
crime_map_europe = folium.Map(location=[54.5260, 15.2551], zoom_start=4)  # Coordinates roughly centered in Europe

# Add the heatmap layer to the map
HeatMap(heatmap_data_europe).add_to(crime_map_europe)

# Display the map
crime_map_europe
```

<h1>Casualty Analysis

```python
# Load the dataset with the correct encoding
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')

# Apply filters to focus on Western Europe
europe_filter = df[df['region_txt'] == 'Western Europe']  # Filter for Western Europe as per the dataset classification

# Group by longitude and latitude and calculate the number of casualties at each location
europe_casualty = europe_filter.groupby(['longitude', 'latitude'])['nkill'].sum().reset_index(name='casualty_count')

# Convert the data into a list format for use in the heatmap
heatmap_data_casualty = europe_casualty[['longitude', 'latitude', 'casualty_count']].values.tolist()

# Create a base map centered on Europe
casualty_map_europe = folium.Map(location=[54.5260, 15.2551], zoom_start=4)  # Coordinates roughly centered in Europe

# Add the heatmap layer to the map
HeatMap(heatmap_data_casualty).add_to(casualty_map_europe)

# Display the map
casualty_map_europe
```

<h1>Nationality Analysis</h1>

```python
# Apply filters to focus on Western Europe
europe_filter = df[df['region_txt'] == 'Western Europe']  # Filter for Western Europe as per the dataset classification

# Handle missing values for longitude, latitude, or country column
europe_filter = europe_filter.dropna(subset=['longitude', 'latitude', 'country_txt'])

# Group by longitude, latitude, and country, and calculate the count of incidents
nationality_analysis = europe_filter.groupby(['longitude', 'latitude', 'country_txt']).size().reset_index(name='incident_count')

# Convert the data into a list format for use in the heatmap
heatmap_data_nationality = nationality_analysis[['longitude', 'latitude', 'incident_count']].values.tolist()

# Create a base map centered on Europe
nationality_map_europe = folium.Map(location=[54.5260, 15.2551], zoom_start=4)

# Add the heatmap layer to the map
HeatMap(heatmap_data_nationality).add_to(nationality_map_europe)

# Display the map
nationality_map_europe
```
